```python
import numpy as np
import pandas as pd
import math

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

import numpy as np


input_file = "buy_computer.csv"


# comma delimited is the default
df = pd.read_csv(input_file, header = 0)
df
```

```python
column_values = df.columns.values
feature_names = column_values[0:4]
feature_names
```

```python
class_names = np.array(['yes','no'])
```

```python
values = df.values
X = values[:,0:4]
X
```

```python
y = values[:,4:]
y 
```

```python


tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file="buy_computer_tree.dot",
        feature_names=feature_names,
        class_names=class_names,
        rounded=True,
        filled=True
)
```
